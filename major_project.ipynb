```python
df=df.rename(columns={df.columns[3]:'ds',df.columns[34]:'y',df.columns[7]:'dt'})
df
```

```python
df_new = df[['ds','y']]
df_new
```

```python
df_new.info()
df_new=df_new.dropna()
df_new
```

```python
import pandas as pd
df_new['ds']=pd.to_datetime(df_new['ds'])
df_new['y']= df_new['y'].astype(int)
df_new
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.plot(df_new['ds'],df_new['y'])
```

```python
df_new2= df_new.head(500)
```

```python
from fbprophet import Prophet
model= Prophet()
model.fit(df_new2)
x= model.make_future_dataframe(periods=30,freq='D')
forcast=model.predict(x)
plt.figure(figsize=(20,5))
#plt.plot(df_new['ds'],df_new['y'])
plt.plot(forcast['ds'],forcast['yhat'])
```
